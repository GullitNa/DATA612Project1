**Introduction**

The recommender system that I plan to build is designed to help both gamers or casual consumers to discover new games they are likely to enjoy, based on their past ratings shown in the following dataset "Project1videogames" and the preferences of similar users.

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/GullitNa/DATA612Project1/main/Project1videogames.csv"
df = pd.read_csv(url)
df.set_index('User', inplace=True)
df.head()
```

After importing the dataset using my Github Repository, the user-item matrix is available to be broken up into separate training and test datasets.

```python
ratings = df.copy()
train=ratings.copy() # training dataset
test=pd.DataFrame(index=ratings.index, columns=ratings.columns) #test dataset
np.random.seed(123)
for user in ratings.index:
    rated_items = ratings.columns[ratings.loc[user].notna()]
    if len(rated_items) >= 2:
        test_items = np.random.choice(rated_items, size=2, replace=False)
        for item in test_items:
            test.at[user, item] = train.at[user, item]
            train.at[user, item] = np.nan

print("Training dataset:")
display(train.head())

print("Test dataset:")
display(test.head())
```

```python
global_mean =train.stack().mean()
print(global_mean)
```

Using the training dataset, I can calculate that the global mean average is 3.5064935064935066 and therefore, I can predict the missing ratings in the test set using the global mean.

```python
global_pred=test.copy()
global_pred=global_pred.fillna(global_mean)
global_pred
```

The following code chunkcs are examples of how I calculate the RMSE for the raw average for both training dataset and the test dataset.

```python
from sklearn.metrics import mean_squared_error

actual_train = train.stack()
predicted_train = pd.DataFrame(global_mean, index=train.index, columns=train.columns).stack()
# MATCH
predicted_train = predicted_train[actual_train.index]
rmse_train = np.sqrt(mean_squared_error(actual_train, predicted_train))
print(rmse_train)
```

```python
actual_test = test.stack()
predicted_test = global_pred.stack()

predicted_test = predicted_test[actual_test.index]
rmse_test = np.sqrt(mean_squared_error(actual_test, predicted_test))
print(rmse_test)
```

Based on these calculations, the test RMSE being 0 means that in this particular instance the model had accurately predicted the results in the test dataset.

Additonal calculations below are the biases of the users and items using my training data:

```python
# User bias
user_bias = train.sub(global_mean).mean(axis=1, skipna=True)
print("User Biases:")
print(user_bias)

# Item bias
item_bias = train.sub(global_mean).mean(axis=0, skipna=True)
print("Item Biases:")
print(item_bias)
```

Based on these biases, User3 has the highest positive bias rating of 0.922078 meaning that they are more willing to rate higher than average. Whereas User4 has the lowest bias as in -1.506494 and aren't as likely to give out much positive ratings.

In terms of items, Fortnite was the highest rated game with 0.743506 above average rating, and Among Us being the lowest rated game with -0.728716 below average rating.

After obtaining the raw average, and both the user and item biases, here is the calculated baseline predictors for every user-item combination:

```python
baseline_predictions = pd.DataFrame(index=train.index, columns=train.columns)

for user in train.index:
    for item in train.columns:
        bu = user_bias.get(user, 0)
        bi = item_bias.get(item, 0)
        baseline_predictions.at[user, item] = global_mean + bu + bi

baseline_predictions
```
